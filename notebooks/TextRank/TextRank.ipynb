```python
import pandas as pd
import numpy as np
from gensim.summarization.summarizer import summarize
import nltk
from rouge_score import rouge_scorer
from sklearn.model_selection import train_test_split
import sklearn
import rouge_score
```

```python
# rouge scorer
scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
```

```python
def try_summarize(x):
    try:
        return summarize(x, ratio=0.1, split=False)
    except:
        return x
```

```python
# Read the test data
X_test = pd.read_csv('../data/interim/X_test.csv')
y_test = pd.read_csv('../data/interim/y_test.csv')
```

```python
summs = X_test['abstracts'].apply(lambda x: try_summarize(x))
```

```python
y_test
```

```python
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores_text_rank = [scorer.score(y_test.titles.values[i], summs[i]) for i in range(y_test.shape[0])]
rouge_scores_text_rank_rouge1 = [x['rouge1'].fmeasure for x in rouge_scores_text_rank]
rouge_scores_text_rank_rouge2 = [x['rouge2'].fmeasure for x in rouge_scores_text_rank]
rouge_scores_text_rank_rougeL = [x['rougeL'].fmeasure for x in rouge_scores_text_rank]
```

```python
bleu_scores_text_rank = [nltk.translate.bleu_score.sentence_bleu(y_test.titles.values[i].split(' '), summs[i].split(' ')) for i in range(y_test.shape[0])]
```

```python
print(f"Rouge1 Scores: {np.mean(rouge_scores_text_rank_rouge1)}")
print(f"Rouge2 Scores: {np.mean(rouge_scores_text_rank_rouge2)}")
print(f"RougeL Scores: {np.mean(rouge_scores_text_rank_rougeL)}")
print(f"BLEU Scores: {np.mean(bleu_scores_text_rank)}")
```
